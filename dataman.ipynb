#### 2.2 top10k and spotify api integration

```python
top10K_clean.isnull().sum()
```

```python
# filling missing artist genres using spotify api
missing = top10K_clean[top10K_clean['Artist Genres'].isnull()]
# some rows contains more than one artist, so we split and explode them
missing['Artist URI(s)'] = missing['Artist URI(s)'].str.split(', ')
missing = missing.explode('Artist URI(s)')
missing.head()
```

```python
# extract the genres from spotify API
artist_URIs = missing['Artist URI(s)']

# Split artist URIs into batches for efficient processing
batch_size = 50  # You can adjust the batch size as needed
batches = [artist_URIs[i:i + batch_size] for i in range(0, len(artist_URIs), batch_size)]

d = {}

# Process batches and retrieve artist genres
for batch in batches:
    artist_infos = sp.artists(batch)
    for artist_info in artist_infos['artists']:
        uri = artist_info['uri']
        genres = artist_info['genres']
        d[uri] = genres

# 'd' contains artist genres for each URI
```

```python
d
```

```python
# most arrays are still empty, which means they are not classified (yet)
# extract only the URIs with an artist genre
d_notnull = {k:v for k,v in d.items() if v != []}
notnull = pd.DataFrame.from_dict(d_notnull, orient='index', columns = ['Artist Genres2'])
notnull = notnull.reset_index(names = 'Artist URIs2')
notnull
```

```python
# now we substitute the artist genres found in the original dataset
top10k_clean2 = top10K_clean.merge(notnull, left_on="Artist URI(s)", right_on="Artist URIs2", how="left")
top10k_clean2["Artist Genres"] = top10k_clean2["Artist Genres2"].fillna(top10k_clean2["Artist Genres"])
top10k_clean2 = top10k_clean2.drop(["Artist URIs2", "Artist Genres2"], axis=1)
top10k_clean2.head()
```

```python
# extract the availabe markets from spotify API
# first we create a new column with all NaN values
top10k_clean2['Country'] = np.nan
top10k_clean2
```

#### 2.3 radio api pre integration

```python
# now we merge the songs with radio API dataset
# first we see that the country format is different for the two datasets, so we add a column with the country code in the radio_stations dataframe
# to perfome this mapping we use a library named 'pycountry'

# Function to map country names to ISO 3166-1 alpha-2 country codes
def map_country_to_alpha2(country_name):
    try:
        country = pycountry.countries.get(name=country_name)
        return country.alpha_2
    except AttributeError:
        return None

# Create a new column 'country_code' by mapping 'country_name' to ISO codes
radio_metadata['country_code'] = radio_metadata['country_name'].apply(map_country_to_alpha2)

# Display the DataFrame
radio_metadata
```

#### 2.5 tracks features

```python
# Batch size for API requests
batch_size = 10

df_tracks = pd.DataFrame(top10k_spotify['Track URI'])

# Function to fetch audio features for a batch of track URIs
def get_audio_features_batch(track_uris):
    audio_features = []
    for i in range(0, len(track_uris), batch_size):
        batch = track_uris[i:i + batch_size]
        features_batch = sp.audio_features(batch)
        audio_features.extend(features_batch)
    return audio_features

# Fetch audio features for all track URIs in the dataframe
audio_features = get_audio_features_batch(df_tracks['Track URI'])

# Convert audio features to a dataframe
df_audio_features = pd.DataFrame(audio_features)

# Merge the audio features dataframe with the original dataframe on the track URI
top10k_features = df_tracks.merge(df_audio_features, left_on='Track URI', right_on='uri', how='left')

# Drop the 'uri' column and keep only the desired audio features
top10k_features = top10k_features.drop(columns=['uri', 'time_signature', 'analysis_url', 'id', 'type', 'mode', 'key'])

top10k_features
```
